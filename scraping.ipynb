```python
import requests
from bs4 import BeautifulSoup   
import pandas as pd
url = "https://books.toscrape.com/"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python

head = soup.find("title")
print(head.get_text(strip=True))
```

```python
titulo =soup.find("div", class_="col-sm-8 h1")
print(titulo.get_text(strip=True))
```

```python
productos = soup.select("article.product_pod")
articulos = {
    "nombre": [],
    "precio": [],
    "imagen_url": []       
}
articulos_list = []

for producto in productos:
    nombre = producto.find("h3").find("a")["title"]
    articulos["nombre"].append(nombre)
    print(nombre)

    precio = producto.find("div" ,class_ = "product_price").find("p").get_text(strip = True)
    articulos["precio"].append(precio)
    print(precio)
    imagen = producto.find("div").find("img")["src"]
    
    imagen_url= url + imagen
    articulos["imagen_url"].append(imagen_url)
    print(imagen_url)
    articulos_list.append({
    "nombre": nombre,
    "precio": precio,
    "imagen_url": imagen_url       
})
```

```python
print(articulos)
print(articulos_list)
```

```python
tabla = pd.DataFrame(articulos_list,index=None, columns=["nombre","precio","imagen_url"])


```

```python
tabla.to_excel("articulos_listar.xlsx", index = False)
```

```python
import requests
from bs4 import BeautifulSoup   
import pandas as pd
url_base = "https://books.toscrape.com/catalogue/page-{}.html"
import time
import random



```

```python
articulos_paginacion=[]

for page in range(1,4):
    url = url_base.format(page)
    response= requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    productos= soup.select("article.product_pod")
    
    for producto in productos:
        nombre = producto.find("h3").find("a")["title"]
        precio = producto.find("div", class_="product_price").find("p").get_text(strip = True)
        imagen = producto.find("div").find("img")["src"]
        imagen_url= url + imagen  
        articulos_paginacion.append({
            "nombre": nombre,
            "precio": precio,
            "imagen_url": imagen_url
        })
time.sleep(1)        
```

```python
tabla = pd.DataFrame(articulos_paginacion, index=None)


tabla.to_excel("articulos_pagina.xlsx", index=False)
```
