```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
url = 'https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29'

response = requests.get(url).text
soup = BeautifulSoup(response,"html.parser")

tables = soup.find_all("tbody")

filas = tables[2].find_all("tr")
paises=[]
for fila in filas:
    columnas = fila.find_all("td")
    if len(columnas)!=0:
        if columnas[0].find("a") is not None and ("—") not in columnas[2]:
            paises.append( {"Country":columnas[0].find("a").get_text(strip=True),
                           "GDP_Millon_Dollars":columnas[2].get_text(strip=True)
                
            })
tabla= pd.DataFrame(paises,index=None)

```

```python
GDP_list =tabla["GDP_Millon_Dollars"].to_list()
GDP_list = [float("".join(x.split(",")))for x in GDP_list]
print(GDP_list)
```
